```python
Q1  - What is the difference between interpreted and compiled languages?

      #Compiled Languages

      1. Code conversion: The programmer writes the code in a high-level language, which is then converted into machine code 
         using a compiler.
      2. Compilation step: The compilation step occurs before the code is executed. The compiler checks the code for errors,
         generates machine code, and creates an executable file.
      3. Execution: The executable file is run directly by the computer's processor, without the need for an interpreter.
      4. Examples: C, C++, and Fortran are examples of compiled languages.

      #Interpreted Languages

     1. Code interpretation: The programmer writes the code in a high-level language, which is then interpreted line-by-line
        by an interpreter at runtime.
     2. No compilation step: There is no separate compilation step. The interpreter reads the code, analyzes it, and executes
        it directly.
     3. Execution: The interpreter executes the code line-by-line, without creating an executable file.
     4. Examples: Python, JavaScript, and Ruby are examples of interpreted languages.

     #Key differences

     1. Compilation: Compiled languages require a compilation step, while interpreted languages do not.
     2. Execution: Compiled languages are executed directly by the processor, while interpreted languages are executed by an interpreter.
     3. Error handling: Compiled languages typically detect errors during the compilation step, while interpreted languages detect
        errors during execution.
     4. Performance: Compiled languages are generally faster than interpreted languages, since they do not require the overhead
        of an interpreter.
```

```python
Q15 -  How does Python's garbage collection system work?
       Python's garbage collection (GC) system is an automated memory management mechanism that deallocates memory no longer used 
       by a program. It helps manage memory effectively by removing objects that are no longer accessible, ensuring that the program
       does not run out of memory due to unused objects.

       Key Components of Python's Garbage Collection-

       #Reference Counting

       Each object in Python maintains a reference count, which tracks how many references point to that object.
       When the reference count drops to zero (i.e., no variable or object points to it), the memory occupied by the object is
       immediately deallocated.

       #Cyclic Garbage Collection

       Python uses a cyclic garbage collector to handle objects that reference each other, forming reference cycles that reference
       counting alone cannot detect.

       #Generational Garbage Collection

        Python’s GC divides objects into three generations: generation 0, generation 1, and generation 2.
        Newly created objects are placed in generation 0.
        Objects that survive garbage collection in generation 0 are promoted to higher generations (1 and 2).
        Higher generations are collected less frequently than lower generations because older objects are more likely to remain in use.

       #Manual Control

       Python provides the gc module for explicit control over garbage collection. For example:
       Disable GC: gc.disable()
       Enable GC: gc.enable()
       Force collection: gc.collect()

```
